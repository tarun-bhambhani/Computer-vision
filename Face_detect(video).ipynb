```python
import cv2
```

```python
x=facehaar=cv2.CascadeClassifier('face.xml')
```

```python
cap=cv2.VideoCapture(0)#starting camera
```

```python
while cap.isOpened():
    status,frame=cap.read()
    face_only =facehaar.detectMultiScale(frame,1.15,5)
    print(face_only)
    for x,y,w,h in face_only:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,200),2)
    
    cv2.imshow('fd',frame)
    if cv2.waitKey(10) & 0xff == ord('q'):
        break

        
cv2.destroyAllWindows()
cap.release()
```
